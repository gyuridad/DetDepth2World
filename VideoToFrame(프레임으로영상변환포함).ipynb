```python
import cv2
import os

# 저장할 폴더 경로
output_dir = "video/지노드론내서읍_Frame"
os.makedirs(output_dir, exist_ok=True)

# 영상 열기
video = cv2.VideoCapture("video/지노드론영상샘플/지노드론내서읍.mp4")

frame_idx = 0

while True:
    ret, frame = video.read()
    if not ret:
        break

    # 저장 파일 이름 만들기
    filename = f"{output_dir}/frame_{frame_idx:05d}.jpg"  # 00001, 00002처럼 저장
    cv2.imwrite(filename, frame)

    frame_idx += 1

video.release()
print(f"✅ 총 {frame_idx}개의 프레임이 {output_dir} 폴더에 저장되었습니다.")
```

```python
import cv2
import os

# 이미지 프레임이 저장된 폴더
frame_folder_name = "yolo_frames"   # "yolo_frames" or "depth_frames"
image_folder = f"./results/{frame_folder_name}"  
# 저장할 비디오 파일 이름
output_video = f"./results/{frame_folder_name}/{frame_folder_name}2video.mp4"

# 프레임들을 정렬된 리스트로 가져오기
images = [img for img in sorted(os.listdir(image_folder)) if img.endswith(".jpg") or img.endswith(".png")]
if not images:
    raise Exception("이미지 프레임이 없습니다.")

# 첫 이미지로 프레임 사이즈 확인
first_frame = cv2.imread(os.path.join(image_folder, images[0]))
height, width, layers = first_frame.shape
size = (width, height)

# 비디오 저장 설정 (코덱: mp4v, FPS: 30)
out = cv2.VideoWriter(output_video, cv2.VideoWriter_fourcc(*'mp4v'), 30, size)

# 이미지들을 비디오로 저장
for image in images:
    frame = cv2.imread(os.path.join(image_folder, image))
    out.write(frame)

out.release()
print(f"✅ 영상 저장 완료: {output_video}")
```
